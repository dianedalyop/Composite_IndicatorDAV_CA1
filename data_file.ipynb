developing a composite indicator which is a single
number that can be used to compare countries, people or cities.
An indicator is a quantitative or qualitative measure derived from a series of observed
facts that can reveal relative positions (e.g. of a country) in a given area. When
evaluated at regular intervals, an indicator can point out the direction of change across
different units and through time.

```python
import pandas as pd

df = pd.read_csv("composite_data.csv")

df_pivot = df.pivot_table(
    index="Country Name",
    columns="Series Name",
    values="2022 [YR2022]",
    aggfunc="first"
)

df_pivot.reset_index(inplace=True)

# removing aggregate rows
df_pivot = df_pivot[~df_pivot["Country Name"].str.contains(
    "Africa|World|income|Europe|Asia|IDA|IBRD|Middle East|North America|Latin America|Caribbean|Arab World|Euro area",
    na=False
)]

# replacing missing markers
df_pivot = df_pivot.replace("..", pd.NA)

# converting only non-country columns to numeric
for col in df_pivot.columns:
    if col != "Country Name":
        df_pivot[col] = pd.to_numeric(df_pivot[col], errors="coerce")

# droping empty country rows
df_pivot = df_pivot.dropna(subset=["Country Name"])

df_pivot.head()
```
